```python
# Importing Matplotlib
import matplotlib.pyplot as plt
# plt is athe alias name for pyplot
import numpy as np
# np is the alias for numpy
import pandas as pd
# pd is the alias for pandas
import seaborn as sns
# seaborn is aliased as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading Dataset
df = pd.read_csv('tips.csv')
# Print the Type of df
print(type(df))
# Show top 5 rows
df.head()  
```

```python
df['day'].unique() # finding unique values in day column
```

# Scatter plot

```python
# Plotting a scatter plot
sns.scatterplot(data=df,x='total_bill',y='tip')
# Setting title for the plot
plt.title('Total Bill vs Tips')
plt.show() # Showing the plot
```

```python
plt.figure(figsize=(8,6)) # setting plot figure size
# Plotting a scatter plot and applying customization
sns.scatterplot(data=df,x='total_bill',y='tip',palette='Set1',hue="day", style="time")
# Setting title for the plot
plt.title('Total Bill vs Tips based on Day')
plt.legend(loc=9) # Setting Location of legend inside the plot
plt.grid() # Showing grid for the plot 
plt.show()# Showing the plot
```

```python
plt.figure(figsize=(8,4)) # setting plot figure size
# Plotting a scatter plot and applying customization
sns.scatterplot(data=df,x='total_bill',y='tip',hue='sex',
                palette='Set1',size='sex')
# Setting title for the plot
plt.title('Total Bill vs Tips based on Sex')
# Setting position and anchoring the legend outside the plot
plt.legend(bbox_to_anchor=(1, 1))
plt.grid() # Showing grid for the plot 
# Showing the plot
plt.show()
```

## Box Plot

```python
plt.boxplot(data=df,x='total_bill')# Plotting a Box plot
plt.title('Total Bill Boxplot') # Setting title for the plot
plt.grid() # Showing grid for the plot 
plt.show() # Showing the plot
```

```python
sns.boxplot(data=df,x='day',y='tip') # Plotting a Box plot
plt.title('Day wise Tips - Boxplot') # Setting title for the plot
plt.grid() # Showing grid for the plot 
plt.show() # Showing the plot
```

```python
sns.boxplot(data=df,x='day',y='tip',hue='sex') # Plotting a Box plot
plt.title('Day wise Tips - Boxplot') # Setting title for the plot
plt.grid() # Showing grid for the plot 
plt.show() # Showing the plot
```

## Violin Plot

```python
sns.violinplot(data=df,x='day',y='tip') # Plotting a Violin Plot
plt.title('Day vs Tips - Violin plot') # Setting title for the plot
plt.show() # Showing the plot
```

```python
plt.figure(figsize=(10,4)) # setting plot figure size
sns.violinplot(data=df,x='day',y='total_bill',hue='smoker')  # Plotting a Violin Plot
plt.title('Day wise Total Bill based on Smoker type - Violin Plot') # Setting title for the plot
plt.show() # Showing the plot
```

```python
plt.figure(figsize=(10,4)) # setting plot figure size
sns.violinplot(data=df,x='day',y='tip',hue='smoker',split=True) # Plotting a Violin Plot
plt.title('Day wise Tips based on Smoker type - Violin plot') # Setting title for the plot
plt.show() # Showing the plot
```
