```python
# Importing Matplotlib
import matplotlib.pyplot as plt
# plt is athe alias name for pyplot
import numpy as np
# np is the alias for numpy
import pandas as pd
# pd is the alias for pandas
import seaborn as sns
# seaborn is aliased as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading Dataset
df = pd.read_csv('tips.csv')
# Print the Type of df
print(type(df))
# Show top 5 rows
df.head()  
```

```python
df.count() # count the total no of entries in the dataframe 
```

```python
df['day'].unique() # finding unique values in day column
```

```python
# Making a new dataframe by grouping the data according to the day column
# Calculating the mean of numeric columns according to group by condition and resetting the index
df_day_mean = df.groupby('day').mean().reset_index()
 # Rounding 'total_bill','tip','size' values
df_day_mean[['total_bill','tip','size']]=round(df_day_mean[['total_bill','tip','size']])
# Converting the float values in 'total_bill','tip','size' to int values
df_day_mean[['total_bill','tip','size']] = df_day_mean[['total_bill','tip','size']].astype(int) 
# Showing the dataframe
df_day_mean
```

```python
df_day_mean.dtypes # Finding the datatypes of all columns
```

Customizing plots

```python
# Setting the figure size for the plot
sns.set(rc={'figure.figsize':(10,8)})
# Setting background for the plot (setting grid style)
sns.set_style('whitegrid')
# setting figure style, font scale and grid line width for the plot
sns.set_context("poster", font_scale = .8, rc={"grid.linewidth": 0.5})
# Plotting a Bar plot
sns.barplot(data=df_day_mean, x='day',y='total_bill')
# Setting title for the plot
plt.title('Day wise Total Bill')
plt.xlabel('Day') # setting x-label
plt.ylabel('Total Bill')  # setting y-label
# Rotating x-label by 45 degrees
plt.xticks(rotation=45)
# Showing the plot
plt.show()
```

```python
# Setting the figure zise for the plot
sns.set(rc={'figure.figsize':(10,8)})
# setting color palette for the plot
sns.set_palette("bright")
# Setting backgroung for the plot (setting grid style)
sns.set_style('darkgrid')
# setting figure style, font scale and gride line width for the plot
sns.set_context("paper", font_scale = 2, rc={"grid.linewidth": 0.5})
# Plotting a Bar plot
sns.barplot(data=df_day_mean, x='day',y='total_bill')
# Setting title for the plot
plt.title('Day wise Total Bill')
plt.xlabel('Day') # setting x-label
plt.ylabel('Total Bill')  # setting y-label
# Rotating x-label by 45 degrees
plt.xticks(rotation=45)
# Showing the plot
plt.show()
```

```python
# Setting the figure zise for the plot
sns.set(rc={'figure.figsize':(10,8)})
# setting color palette for the plot
sns.set_palette("magma")
# Setting backgroung for the plot (setting grid style)
sns.set_style('darkgrid')
# setting figure style, font scale and gride line width for the plot
sns.set_context("paper", font_scale = 2, rc={"grid.linewidth": 0.5})
# Plotting a Bar plot
sns.barplot(data=df_day_mean, x='day',y='total_bill')
# Setting title for the plot
plt.title('Day wise Total Bill')
plt.xlabel('Day') # setting x-label
plt.ylabel('Total Bill')  # setting y-label
# Rotating x-label by 45 degrees
plt.xticks(rotation=45)
# Showing the plot
plt.show()
```

```python
# Making a new dataframe by grouping the data according to the 'day','time' and sex column
# Calculating the sum of numeric columns according to group by condition and resetting the index
df_day_sum = df.groupby(['day','time','sex']).sum().reset_index()
# Converting the float values in 'total_bill','tip','size' to int values
df_day_sum[['total_bill','tip','size']] = df_day_sum[['total_bill','tip','size']].astype(int) 
# Showing the dataframe
df_day_sum
```

```python
# Setting the figure size for the plot
sns.set(rc={'figure.figsize':(10,8)})
# setting color palette for the plot
sns.set_palette("husl")
sns.despine() # automatically take away the top and right spines
# Setting background for the plot (setting grid style)
sns.set_style('darkgrid')
# setting figure style, font scale and grid line width for the plot
sns.set_context("paper", font_scale = 2, rc={"grid.linewidth": 0.5})
# Plotting a line plot
sns.lineplot(data=df_day_sum, x='day',y= 'tip',hue='sex',style='sex',markers=True)
# Setting title for the plot
plt.title('Day wise Total Bill')
plt.xlabel('Day') # setting x-label
plt.ylabel('Total Bill')  # setting y-label
# Rotating x-label by 45 degrees
plt.xticks(rotation=45)
# Showing the plot
plt.show()
```

Plotting Multiple Plot using FacetGrid

```python
# Setting background for the plot (setting grid style)
sns.set_style('darkgrid')
# setting figure style for the plot
sns.set_context("notebook")
# Defining facetgrid according to the required column
sea = sns.FacetGrid(df_day_sum, col = "sex")
# mapping the facetgrid with the desired plot
sea.map(sns.scatterplot, "day", "total_bill")
# Showing the plot
plt.show()
```
