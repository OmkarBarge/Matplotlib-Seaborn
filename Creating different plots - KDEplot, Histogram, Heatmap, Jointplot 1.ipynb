```python
# Importing Matplotlib
import matplotlib.pyplot as plt
# plt is athe alias name for pyplot
import numpy as np
# np is the alias for numpy
import pandas as pd
# pd is the alias for pandas
import seaborn as sns
# seaborn is aliased as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading Dataset
df = pd.read_csv('tips.csv')
# Print the Type of df
print(type(df))
# Show top 5 rows
df.head()  
```

```python
df['day'].unique() # finding unique values in day column
```

## KDE Plot

```python
sns.kdeplot(data=df,x='total_bill') # Plotting a KDE plot
plt.title('KDE plot for Total Bill') # Setting title for the plot
plt.show() # Showing the plot
```

```python
plt.figure(figsize=(8,4)) # setting plot figure size
# Plotting a KDE plot
sns.kdeplot(data=df,x='tip',hue='time',palette='bright')
# Setting title for the plot
plt.title('KDE plt for Tips for Dinner and Lunch time')
plt.show() # Showing the plot
```

## Histogram

```python
sns.histplot(data=df,x='total_bill') # Plotting a Histogram
plt.title('Histplot for Total Bill') # Setting title for the plot
plt.show() # Showing the plot
```

```python
plt.figure(figsize=(8,4)) # setting plot figure size
# Plotting a Histogram
sns.histplot(data=df,x='total_bill',hue='smoker',palette='bright')
# Setting title for the plot
plt.title('Histogram for Total Bill based on Smoker type')
plt.show() # Showing the plot
```

## Heatmap

```python
corr = df.corr() # Finding correlation in data
corr # Showing correlation table
```

```python
plt.figure(figsize=(8,4)) # setting plot figure size
# Plotting a Heatmap and visualizing the correlation
sns.heatmap(data=corr,annot=True,cmap='RdBu')
plt.show() #Showing the plot
```

## Jointplot

```python
sns.jointplot(data=df, x='day',y='total_bill') # Plotting a Jointplot
plt.show() # Showing the plot
```

```python
# Plotting a Jointplot
sns.jointplot(data=df, x='total_bill',y='tip',hue='time',palette='bright',kind='hist')
plt.show() # Showing the plot
```

```python
# Plotting a Jointplot
sns.jointplot(data=df, x='total_bill',y='tip',hue='time',palette='bright',kind='kde')
plt.show() # Showing the plot
```

```python
# Plotting a Jointplot
sns.jointplot(data=df, x='total_bill',y='tip',palette='bright',kind='hex',color='maroon')
plt.show() # Showing the plot
```
