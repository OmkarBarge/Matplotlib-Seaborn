```python
# Importing Matplotlib
import matplotlib.pyplot as plt
# plt is athe alias name for pyplot
import numpy as np
# np is the alias for numpy
# importing Image from the Pillow
from PIL import Image

import warnings
warnings.filterwarnings("ignore")
```

Displaying Image in colored mode

```python
w1 = Image.open('cat_1.jpg') #Opening the image
plt.imshow(w1)  # imshow - image show
plt.show() # Showing the plot
```

Displaying Image in GrayScale Mode

```python
w2 = Image.open('cat_2.jpg') #Opening the image
gray_img = w2.convert("L")  # L implies read in grayscale mode
plt.imshow(gray_img, cmap='gray') # Assigning color map as 'gray'
plt.show()# Showing the plot
```

```python
plt.imshow(gray_img) # Plotting image without color map
```

```python
plt.imshow(gray_img,cmap ='hot')  # Assigning color map as 'hot'
```

```python
plt.imshow(gray_img,cmap ='binary') # Assigning color map as 'binary'
```

```python
dog = Image.open('dog_1.jpg') #Opening the image
newsize = (800, 400) # defining new size
resized_image = dog.resize(newsize) # resizing the image
plt.axis("off") # Hide the axes of the plot by setting  plt.axis() as 'off'
plt.imshow(resized_image)# imshow - image show
plt.show() # showing the plot
```

```python
# Plotting Multiple Plot using Subplots() function
# Plotting 4 plots in a single plot having 2 rows and 2 columns 
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(12,10))  # nrows=2,ncols=2
# ax1 is row1,col1, ax2 is row1,col2
# ax3 is row2,col1, ax4 is row2,col2

# ax1 is row1,col1
w1 = Image.open('cat_1.jpg') #Opening the image
ax1.imshow(w1)  # imshow - image show

# ax1 is row1,col2
w2 = Image.open('cat_2.jpg') #Opening the image
ax2.imshow(w2)  # imshow - image show

# ax1 is row2,col1
w3 = Image.open('dog_1.jpg') #Opening the image
ax3.imshow(w3)  # imshow - image show

# ax1 is row2,col2
w4 = Image.open('dog_2.jpeg') # Importing and opening .jpeg file image
ax4.imshow(w4)  # imshow - image show

plt.show() # showing the plot
```
