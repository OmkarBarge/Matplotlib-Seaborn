### Bar Chart
1) The graph represents categories on one axis and a continuous value in the other.<br>
2) It can be used to compare different categories of data<br>

#### Syntax
<b>plt.bar(x,y,color,label,edgecolor)</b><br>
x - data(column) to be represeted on the x-axis<br>
y - data(column) to be represeted on the y-axis<br>
color - Used to specify color of the bar<br>
edgecolor - Used to define te color of the edge of the bar chart<br>

#### Q) Plot year wise average Runtime of top 7 movies on a bar chart

```python
q1 = d1.groupby('title_year')['Runtime'].mean().reset_index()
# q1 = q1.sort_values(by='title_year',ascending=True).head(7)
q1
```

```python
plt.bar(q1['title_year'],q1['Runtime'])
```

### Graph Customization

### 1) Add xlabel, ylabel, title and show

a) <b>plt.xlabel()</b><br>
It is used to provide label to the x-axis.

b) <b>plt.ylabel()</b><br>
It is used to provide label to the y-axis.

c) <b>plt.title()</b><br>
It is used to provide title to the chart

d) <b>plt.show()</b><br>
This will ensure that all of the figures are shown and this function immediately returns.
It gets rid of the gibberish data displayed above the figure

```python
plt.bar(q1['title_year'],q1['Runtime'])
plt.title('Year wise avg Runtime',fontsize=15,color='orange')
plt.xlabel('Title_Year')
plt.ylabel('RunTime')
plt.show()
```

### 2) Add color and edge color to the bars

e) <b>color</b><br>
It is an attribute used to define the color of the bars in bar chart.<br>
A single color can be chosen or a list of colors can be passed as  an argument

f) <b>edgecolor</b><br>
It is an attribute used to define the edgecolor of the bars in bar chart

```python
plt.bar(q1['title_year'],q1['Runtime'],color='orange',edgecolor='black')
plt.xlabel('Title_Year')
plt.ylabel('RunTime')
plt.title('Year wise mean Runtime')
plt.show()
```

```python
plt.bar(q1['title_year'],q1['Runtime'],
        color=['orange','lightblue','violet','lightgreen','brown','tan','yellow']
        ,edgecolor='black')
plt.title('Year wise avg Runtime',fontsize=15,color='maroon')
plt.xlabel('Title_Year')
plt.ylabel('RunTime')
plt.show()
```

#### 2) Plot genre_1 wise Mean of budget and Gross on bar chart

Converting Budget and Gross from dollars to Millon dollars

```python
d1["Gross"] = d1["Gross"]/1000000
d1["budget"] = d1["budget"]/1000000
d1.head()
```

```python
d1.columns
```

```python
q2 = d1.groupby('genre_1')[['budget','Gross']].mean()
print(type(q2))
q2.index.names = ['Genre']
q2
```

### 2) Add label and legend
g) <b>label</b><br>
It is a parameter of plt.bar() function. It is used to assign a distinct label to each of the categories

h) <b>plt.legend()</b><br>
legend ensures that the label parameter addaed gets reflected in the chart

<b>alpha</b><br>
It is parameter of bar chart used to set the transparency<br>

```python
print(q2.index)
# x axis is having index
```

```python
plt.bar(q2.index,q2['budget'],label='Budget',color='blue')
plt.bar(q2.index,q2['Gross'],label='Gross',color='tan')
plt.xlabel('Genre')
plt.title('Genre wise Mean Budget and Gross')
plt.legend(loc=0)  # loc = 0,1,2,....10
plt.show()
```

```python
plt.bar(q2.index,q2['Gross'],label='Gross',color='tan')
plt.bar(q2.index,q2['budget'],label='Budget',color='blue')
plt.xlabel('Genre')
plt.title('Genre wise Mean Budget and Gross')
plt.legend(loc=0)  # loc = 0,1,2,....10
plt.show()
```

### 3) Add xticks, yticks, alpha

i) <b>plt.xticks()</b><br>
1) It can be used to rotate the label in case they are overlapping on the x-axis values<br>
2) It can also be be used to assign some alias names to the labels on x-axis, i.e It can be used to change the axis scale and ticks<br>

j) <b>plt.yticks()</b><br>
1) It can be used to rotate the label in case they are overlapping on the y-axis values<br>
2) It can also be be used to assign some alias names to the labels on y-axis<br>

k) <b>alpha</b><br>
It is used to set transparency. Used to distinguish the bars using transparencey when they overlap

```python
plt.bar(q2.index,q2['budget'],label='Budget',color='blue')
plt.bar(q2.index,q2['Gross'],label='Gross',color='maroon',alpha=0.5)
plt.xlabel('Genre')
plt.title('Genre wise Mean Budget and Gross')
plt.legend()
plt.xticks(rotation=30)
# plt.yticks(rotation=45)
plt.show()
```

#### 3) Plot actor_1_name  based and their mean of fb_likes for the top 7 actors with highest fb_likes on a bar chart

```python
q3 = d1.groupby('actor_1_name')['actor_1_facebook_likes'].mean().reset_index()
q3 = q3.sort_values(by='actor_1_facebook_likes',ascending=False).head(7)
q3.columns = ['Actor','Fb_Likes']
q3
```

```python
plt.bar(q3['Actor'],q3['Fb_Likes'],color='maroon',label='Fb_likes')
plt.xlabel('Actor')
plt.ylabel('FB_Likes')
plt.title('Actor wise Fb_Likes')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

### 4) Adding xlim, ylim, figure, savefig
l) <b>plt.xlim()</b><br>
It is used to provide range of values for the x-axis.

m) <b>plt.ylim()</b><br>
It is used to provide range of values for the y-axis.

n) <b>plt.figure(figsize)</b><br>
It is used to change the figure size<br>

o) <b>plt.savefig()</b><br>
It is used to save the figure<br>

```python
plt.figure(figsize=(8,5))
plt.bar(q3['Actor'],q3['Fb_Likes'],color='violet',label='Fb_likes')
plt.xlabel('Actor')
plt.ylabel('FB_Likes')
plt.title('Actor wise Fb_Likes')
plt.xticks(rotation=90)
plt.legend()
plt.ylim(0,45000)
plt.savefig('Actor_vs_Fb_likes_14Aug.png')
plt.show()
```

### 5) Adding text, annotation
p) <b>plt.text(x,y,s,color,fontsize)</b><br>
It is used to add text in the chart

q) <b>plt.annotate()</b><br>
It is used add annotation to the chart<br>


#### Adding Text

```python
plt.figure(figsize=(8,5))
plt.bar(q3['Actor'],q3['Fb_Likes'],color='tan',label='Fb_likes')
plt.text(1,15000,'Actor and their FB Likes',color='blue',fontsize=18)
plt.xlabel('Actor')
plt.ylabel('FB_Likes')
plt.title('Actor wise Fb_Likes')
plt.xticks(rotation=60)
plt.legend()
plt.show()
```

#### Adding Annotations

```python
s = list(range(7))  # s = [0,1,2,3,4,5,6]
plt.bar(s,q3['Fb_Likes'],color='orange',label='Fb_likes',edgecolor='black')
plt.ylim(0,45000)
for idx,val in enumerate(q3['Fb_Likes']):  # generates additional counter (idx)
    plt.annotate(xy=[s[idx]-0.4,int(val)+1000],s=int(val))
plt.xticks(s,q3['Actor'],rotation=90)
plt.legend()
plt.show()
```

#### Q) Plot mean Budget vs mean Gross on Bar chart for different years.  Represent Budget and Gross on 2 different axis

```python
q4 = d1.groupby('title_year')[['budget','Gross']].mean()
q4
```

```python
fig = plt.figure() 

ax = fig.add_subplot(111) 
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

width = 0.4

q4['budget'].plot(kind='bar', color='orange', ax=ax, width=width, position=1)
q4['Gross'].plot(kind='bar', color='blue', ax=ax2, width=width, position=0)

ax.set_ylabel('Budget')
ax2.set_ylabel('Profit')
plt.show()
```
